```python
import requests
```

```python
def send_search_request(url, query, top_k=10, return_type='dataframe'):

    data = {"top_k":top_k, "mode": "search", "data": [query]}
    
    response = requests.post(url, json = data)
    if response.status_code == requests.codes.ok:
        matches = response.json()['search']['docs'][0]['matches']
        match_info = []
        for match in matches:            
            match_info.append((match['text'].replace("\n",""),(1-match['score']['value'])))
            
        if return_type == 'dataframe':
            import pandas as pd
            return pd.DataFrame(match_info, columns=['match', 'score'])
                
             
        return match_info
    return "Connection Failed!"
```
